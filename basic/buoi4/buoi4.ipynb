Task 1: Sắp xếp và in ra 10 thành phố có dân số lớn nhất & 10 thành phố có dân số nhỏ nhấ

```python
# Task 1
df_sort_by_population = df
df_sort_by_population["Population"] = df_sort_by_population["Population"].replace(
    ",", "", regex=True
)
df_sort_by_population["Population"] = df_sort_by_population["Population"].astype(
    "int32"
)
df_sort_by_population = df.sort_values(by="Population", ascending=False)
```

#task 2 In ra tên các quốc gia có tối thiểu 3 thành phố trong danh sách này


```python
df_countrys = df
result = df_countrys["Country"].value_counts().to_dict()
result = [(k.replace("\xa0", ""), v) for (k, v) in result.items() if v >= 3]
result
```

Task 4: In ra các thành phố có dân số & diện tích đều nằm trong Top 20

```python
df_top_20 = df
df_top_20["Population"] = df_top_20["Population"].replace(",", "", regex=True)
df_top_20["Population"] = df_top_20["Population"].astype("int32")
df_top_20 = df_top_20.sort_values(["Population", "Area KM2"], ascending=False).head(20)[
    "City"
]
df_top_20
```

Task 5: Thống kê mật độ dân số theo quốc gia

```python
df_by_country = df
df_by_country["Density KM2"] = df_by_country["Density KM2"].replace(",", "", regex=True)
df_by_country["Density  M2"] = df_by_country["Density  M2"].replace(",", "", regex=True)
as_type_obj = {
    "Density KM2": "float32",
    "Density  M2": "float32",
}

df_by_country = df_by_country.astype(as_type_obj, errors="ignore")
df_by_country.groupby("Country")["Density KM2", "Density  M2"].mean()
```

Task 6: Thống kê các thành phố có dân số lớn nhất của từng quốc gia (chỉ tính của những quốc gia có 2
thành phố xuất hiện trở lên trong bảng)

```python
df_result = df
df_result["Population"] = df_result["Population"].replace(",", "", regex=True)
df_result["Population"] = df_result["Population"].astype("int32")
```

```python
# danh sach quoc gia co thanh pho xuat hien it nhat 2 lan
city_least_two_times = df_result["Country"].value_counts().to_dict()
city_least_two_times = [k for (k, v) in city_least_two_times.items() if v >= 2]
# data frame co quoc gia co thanh pho xuat hien it nhat 2 lan
df_result = df_result[df_result["Country"].isin(city_least_two_times)]
```

```python
# lay ra thanh pho co dan so lon nhat theo tung quoc gia
from pandas import DataFrame


def get_max_item(item: DataFrame) -> DataFrame:
    # print(item)
    # print("-------------------")
    return item.nlargest(1, "Population")


largest_cities = df_result.groupby("Country").apply(
    lambda x: x.nlargest(1, "Population")
)
# largest_cities = df_result.groupby('Country').apply(get_max_item)
largest_cities
```

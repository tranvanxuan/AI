```python
import pandas as pd
```

```python
df = pd.read_csv('./Basic_Class_Buoi_5/project_data.csv')
df
```

Task 1: Thống kê số lượng khách hàng theo học vấn (High school,
master,...)

```python
df_by_education_level = df.copy(deep=True)
df_by_education_level.groupby('educational_level').size()
```

Task 2: In ra 20 khách hàng có thu nhập cá nhân (annual_income) cao nhất

```python
df_top20_annual_income = df.copy(deep=True)
annual_income_is_not_null = df['annual_income'].notnull()
df_top20_annual_income[annual_income_is_not_null].sort_values(by="annual_income", ascending=False).head(20)[["customer_id", "annual_income", "educational_level"]]
```

Task 3: In ra những khách hàng sinh sau năm 1960 và thu nhập cá nhân
trên $50,000/năm

```python
df_by_time = df.copy(deep=True)
df_by_time = df_by_time[(df_by_time["year_of_birth"]> 1960) & (df_by_time["annual_income"]>50000)]
df_by_time
```

Task 4: Kết hợp điều kiện của Task 2 & 3

```python
annual_income_is_not_null =  df_by_time['annual_income'].notnull()
df_by_time[annual_income_is_not_null].sort_values(by="annual_income", ascending=False).head(20)[["customer_id", "annual_income", "year_of_birth", "educational_level"]]
```

```python
df_by_time[annual_income_is_not_null].sort_values(by="annual_income", ascending=False).tail(20)[["customer_id", "annual_income", "year_of_birth", "educational_level"]]
```

Task 5: In ra những khách hàng có tình trạng hôn nhân là đã kết hôn or đã
ly hôn

```python
df_marital_status = df.copy(deep=True)
df_marital_status[(df_marital_status["marital_status"] == "Divorced") | (df_marital_status["marital_status"] == "Married")][["customer_id", "marital_status"]]
```

Task 6: Tính mức thu nhập trung bình theo trình độ học vấn

```python
df_annual_income_mean = df.copy(deep=True)
df_annual_income_mean.groupby("educational_level")["annual_income"].mean()
```

Task 7. Tính mức thu nhập trung bình theo trình độ học vấn + tình trạng
hôn nhân

```python
df_annual_income_marital_status_mean = df.copy(deep=True)
df_annual_income_mean.groupby(["educational_level", "marital_status"])["annual_income"].mean()
```
